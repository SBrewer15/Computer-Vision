```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('elephant.jpg')
```

```python
plt.figure(figsize=(10,10))
plt.axis("off")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
hsv_img = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
# define range of PURPLE color in HSV
lower_clr = np.array([15,100,100])
upper_clr = np.array([45,255,255])
```

```python
# Use inRange to capture only the values between lower & upper_blue
mask = cv2.inRange(hsv_img, lower_clr, upper_clr)
```

```python
plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(mask, cmap='gray')
plt.show()
```

```python
# Perform Bitwise AND on mask and our original frame
res = cv2.bitwise_and(image, image, mask=mask)
```

```python
plt.figure(figsize=(10,10))
plt.axis("off")
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
plt.show()
```
